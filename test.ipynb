```python
import glob
import pandas as pd
```

```python
glob.glob("*html")
```

```python
with open(glob.glob("*html")[0],'r') as f:
    html=f.read()
dfs=pd.read_html(html)
del_col=['授業担当教員','推奨  ▼/▲','科目コード  ▼/▲',r"('内訳', 'R◎')","('内訳', 'A○')","('内訳', 'B○')","('内訳', 'C○')","('内訳', 'D○')","('内訳', 'L(選択)')","('内訳', '標準学修課程外')" ]
for i,df in enumerate(dfs):
    dfs[i]=df.replace('19B30221','ALTforID',regex=True)
    dfs[i].columns=map(lambda x: str(x),dfs[i].columns)
    dfs[i].columns=dfs[i].columns.str.replace(r"[(')]","_")
    for col in del_col:
        if col in dfs[i].columns.tolist():
            dfs[i]=dfs[i].drop(columns=col,)
```

```python
concat_df=pd.DataFrame()
import copy
for df in dfs:
    concat_df=pd.concat([concat_df,df])
concat_df_cpy=copy.copy(concat_df)
```

```python
concat_df.drop(columns=concat_df_cpy.columns[9:16].tolist(),errors='ignore',inplace=True)
concat_df.reset_index(inplace=True,drop=True)
concat_df.drop(index=range(5),inplace=True)
concat_df['科目区分']=concat_df['科目区分'].fillna(concat_df['__科目区分_, _科目区分__'])
concat_df.drop(columns="__科目区分_, _科目区分__",inplace=True)
concat_df['修得済単位']=concat_df['修得済単位'].fillna(concat_df['__修得済単位_, _修得済単位__'])
concat_df.drop(columns="__修得済単位_, _修得済単位__",inplace=True)
```
